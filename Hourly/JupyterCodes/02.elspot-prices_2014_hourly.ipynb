```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
# Use seaborn style defaults and set the default figure size
sns.set(rc={'figure.figsize':(20, 10)})
```

```python
df= pd.read_csv('C:/Users/simhaj-local/Documents/GitHub/NordPool_Price/Hourly/CSVFiles/02.elspot-prices_2014_hourly.csv', header=2)
```

```python
df['DateAndTime'] = pd.date_range('1/1/2014', periods=8761, freq='H')
```

```python
df['SE1'].plot(linewidth=1.5, marker = '.', linestyle = '--')
df['SE2'].plot(linewidth=1.5, marker='o', linestyle='-')
df['SE3'].plot(linewidth=1.5, marker = 'v', linestyle='-')
df['SE4'].plot(linewidth=1.5, marker = 'd')
plt.title('2014 Hourly Price Change')
plt.xlabel('Time')
plt.ylabel('Amount')
plt.legend()
```
