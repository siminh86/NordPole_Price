```python
# Open the data 
precip_file = "C:/Users/simhaj-local/Documents/GitHub/NordPole_Price/Hourly/CSVFiles/06.elspot-prices_2018_hourly - Copy.csv"
precip_boulder = pd.read_csv(precip_file,
                             na_values=[0],
                             parse_dates=['Hours'], 
                             index_col = 'Hours')
precip_boulder.head()
```

```python
# Plot the data 
fig, ax = plt.subplots(figsize = (30,15))
ax.scatter(precip_boulder.index, 
           precip_boulder["Sell-Buy"], 
           color = 'purple')
ax.set(xlabel='Hours', ylabel='Volume Difference',
       title="Hourly price change 2018")
plt.show()
```

```python
# Plot the data 
fig, ax = plt.subplots(figsize = (30,15))
ax.scatter(precip_boulder.index, 
           precip_boulder["Load Price"], 
           color = 'Green')
ax.set(xlabel='Hours', ylabel='Load Difference Price',
       title="Hourly price change 2018")
plt.show()
```
