```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
# Use seaborn style defaults and set the default figure size
sns.set(rc={'figure.figsize':(20, 10)})
```

```python
df= pd.read_csv('C:/Users/simhaj-local/Documents/GitHub/NordPole_Price/Daily/csvFiles/06.elspot-prices_2018_daily_sek.csv', encoding = 'unicode_escape',header=0)
```

```python
df['DateAndTime'] = pd.date_range('1/1/2018', periods=365, freq='D')
```

```python
df['SE1'].plot(linewidth=1.5, marker = '.', linestyle = '--')
df['SE2'].plot(linewidth=1.5, marker='o', linestyle='-')
df['SE3'].plot(linewidth=1.5, marker = 'v', linestyle='-')
df['SE4'].plot(linewidth=1.5, marker = 'd')
plt.title('2018 Daily Price Change')
plt.xlabel('Date')
plt.ylabel('Amount')
plt.legend()
```
