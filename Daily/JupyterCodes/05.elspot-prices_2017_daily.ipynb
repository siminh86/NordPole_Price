```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
# Use seaborn style defaults and set the default figure size
sns.set(rc={'figure.figsize':(30, 20)})
```

```python
df= pd.read_csv('C:/Users/simhaj-local/Documents/GitHub/NordPool_Price/Daily/csvFiles/05.elspot-prices_2017_daily_sek.csv',encoding = 'unicode_escape', header=0)
```

```python
df['DateAndTime'] = pd.date_range('1/1/2017', periods=365, freq='D')
```

```python
df['SE1'].plot(linewidth=1.5, marker = '.', linestyle = '--')
df['SE2'].plot(linewidth=1.5, marker='o', linestyle='-')
df['SE3'].plot(linewidth=1.5, marker = 'v', linestyle='-')
df['SE4'].plot(linewidth=1.5, marker = 'd')
plt.title('2017 Daily Price Change')
plt.xlabel('Date')
plt.ylabel('Amount')
plt.legend()
```
