```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
# Use seaborn style defaults and set the default figure size
sns.set(rc={'figure.figsize':(20, 10)})
```

```python
df = pd.read_csv('C:/Users/simhaj-local/Documents/GitHub/NordPool_Price/Monthly/CSVFiles/06.elspot-prices_2017_monthly.csv',header = 2)
```

```python
df['DateAndTime'] = pd.date_range('1/1/2017', periods=12, freq='M')
```

```python
df['SE1'].plot(linewidth=1.5, marker = '.', linestyle = '--')
df['SE2'].plot(linewidth=1.5, marker='o', linestyle='-')
df['SE3'].plot(linewidth=1.5, marker = 'v', linestyle='-')
df['SE4'].plot(linewidth=1.5, marker = 'd')
plt.title('2017 Monthly Price Change')
plt.xlabel('12 months in 2017')
plt.ylabel('Amount')
plt.legend()
```
