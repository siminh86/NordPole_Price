```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
# Use seaborn style defaults and set the default figure size
sns.set(rc={'figure.figsize':(20, 10)})
```

```python
df_2012 = pd.read_csv('C:/Users/simhaj-local/Documents/GitHub/NordPool_Price/Monthly/CSVFiles/01.elspot-prices_2012_monthly.csv',header = 2)
```

```python
df_2012.shape
```

```python
df_2012['DateAndTime'] = pd.date_range('1/1/2012', periods=12, freq='M')
```

```python
df_2012.set_index('DateAndTime')
```

```python
df_2012['SE1'].plot(linewidth=1.5, marker = '.', linestyle = '--')
df_2012['SE2'].plot(linewidth=1.5, marker='o', linestyle='-')
df_2012['SE3'].plot(linewidth=1.5, marker = 'v', linestyle='-')
df_2012['SE4'].plot(linewidth=1.5, marker = 'd')
plt.title('2012 Monthly Price Change')
plt.xlabel('12 months in 2012')
plt.ylabel('Amount')
plt.legend()
```
